```python
cursor = cnx.cursor()
query = ("SELECT text FROM review limit 60000")
cursor.execute(query) 
results = cursor.fetchall()
```

```python
with open('review_text_meta.csv','w') as f1:
    writer=csv.writer(f1, delimiter='\t',lineterminator='\n',)
    count = 0
    for row in results:
        if (count %10000 == 0):
            print(count)
        count += 1
        row = word_tokenize(row[0])
        #Number of tokens
        num_token = (float)(len(row))
        row_pos_tag = pos_tag(row)
        #Number of nouns, verbs, adjectives, and adverbs
        num_noun = len([token for token,pos in row_pos_tag if pos.startswith('N')])
        num_verb = len([token for token,pos in row_pos_tag if pos.startswith('V')])
        num_adj = len([token for token,pos in row_pos_tag if pos.startswith('J')])
        num_adverb = len([token for token,pos in row_pos_tag if pos.startswith('R')])
        #Number of words starting with uppercase
        num_upper = 0
        #Number of ! and ?
        num_exclaimation = 0
        num_question = 0
        for token in row:
            if token[0].isupper():
                num_upper += 1
            elif token == "!":
                num_exclaimation += 1
                num_question += 1
        #Compute the ratio of each feature and total number of tokens
        towrite = [num_token, num_noun/num_token, num_verb/num_token, num_adj/num_token, num_adverb/num_token,
                  num_exclaimation/num_token, num_question/num_token]
        writer.writerow(towrite)
                
print("Done extracting meta data features from the review text")
```
